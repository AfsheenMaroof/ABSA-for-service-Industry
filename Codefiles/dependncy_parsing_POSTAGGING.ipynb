```python
import spacy
import pandas as pd
import string

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Load your Excel file
df = pd.read_excel('/content/Final Dataset for analysis .xlsx')

# Assuming the review text is in a column named 'Sentences'
review_texts = df['Sentences']

# Define the preprocess_text function
def preprocess_text(text):
    text = text.translate(str.maketrans('', '', string.punctuation)).lower()
    return text

# Initialize an empty list to store the parsed reviews
parsed_reviews = []  # For POS tagging and dependency parsing together

# Process each review text and perform both tasks
for review_text in review_texts:
    # Preprocess the text
    preprocessed_review = preprocess_text(review_text)

    # Parse the preprocessed text with spaCy
    doc = nlp(preprocessed_review)

    # Store POS tagging and dependency information for the sentence
    sentence_info = []
    for token in doc:
        word = token.text
        root = token.head.text if token.dep_ != 'ROOT' else 'ROOT'
        pos_tag = token.pos_
        dep_label = token.dep_
        sentence_info.append((word, root, pos_tag, dep_label))

    # Append the parsed review to the list
    parsed_reviews.append(sentence_info)

# Now, parsed_reviews contains both POS tagging and dependency parsing results for each sentence
```

```python
# Print word, root, POS tag, and dependency label for each token in the parsed reviews
for i, sentence_info in enumerate(parsed_reviews):
    print(f"Review {i + 1} - POS Tagging and Dependency Parsing:")
    for word, root, pos_tag, dep_label in sentence_info:
        print(f"Word: {word}, Root: {root}, POS Tag: {pos_tag}, Dependency Label: {dep_label}")
    print("\n")
```

```python
# Print word ID, parent ID, word, root, POS tag, and dependency label for each token in the parsed reviews
for i, sentence_info in enumerate(parsed_reviews):
    print(f"Review {i + 1} - POS Tagging and Dependency Parsing:")
    for word_id, (word, root, pos_tag, dep_label) in enumerate(sentence_info):
        parent_id = next((idx for idx, (_, _, _, parent) in enumerate(sentence_info) if parent == word), "ROOT")
        print(f"Word ID: {word_id}, Parent ID: {parent_id}, Word: {word}, Root: {root}, POS Tag: {pos_tag}, Dependency Label: {dep_label}")
    print("\n")
```

Storing in data frame

```python
import pandas as pd

# Create an empty DataFrame
parsed_reviews_df = pd.DataFrame(columns=["Sentence ID", "Word ID", "Parent ID", "Word", "Root", "POS Tag", "Dependency Label"])

# Create an empty list to hold token_data
token_data_list = []

for i, sentence_info in enumerate(parsed_reviews):
    for word_id, (word, root, pos_tag, dep_label) in enumerate(sentence_info):
        parent_id = next((idx for idx, (_, _, _, parent) in enumerate(sentence_info) if parent == word), "ROOT")
        token_data = {
            "Sentence ID": i,
            "Word ID": word_id,
            "Parent ID": parent_id,
            "Word": word,
            "Root": root,
            "POS Tag": pos_tag,
            "Dependency Label": dep_label
        }
        token_data_list.append(token_data)

# Concatenate the list of token_data into the DataFrame
parsed_reviews_df = pd.concat([parsed_reviews_df, pd.DataFrame(token_data_list)])

```

```python
# Save the parsed data to an Excel file
output_file_path = 'parsed_reviews.xlsx'  # Specify the desired file path

# Use the to_excel() method to save the DataFrame to an Excel file
parsed_reviews_df.to_excel(output_file_path, index=False)

print(f"Parsed data saved to {output_file_path}")
```
